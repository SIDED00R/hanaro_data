### 1-1. status code (상태 코드)
- http의 상태를 표현하는 코드
- 200: 요청한 정보에 대한 반환 성공
- 404: 페이지를 찾을 수 없음
- 500: 서버의 에러 등에 의한 url에 대한 정보를 제공받을 수 없는 상황
- 400: 서버에 잘못된 요청을 한 경우
- 403: 서버에 의해 해당 유저의 접근이 차단 됨

### 1-3. robots.txt

- 웹사이트가 웹크롤러에게 자신의 사이트를 어떻게 크롤링 해야 하는지 지정해주는 역할
- 보안 목적이 아닌 여러 검색엔진 내 크롤러에게 어떤 정보를 줄지 제공하는 역할
- 검색 엔진(search engine): 사용자가 원하는 정보를 입력하면, 그에 맞는 웹페이지나 정보를 찾아주는 프로그램
  - 1. 크롤링(crawling): 검색엔진은 웹크롤러라고 불리는 프로그램을 사용하여 수많은 웹페이지를 자동으로 탐색
   
  - 2. 인덱싱(indexing): 수집한 웹페이지의 내용을 분석하여 색인화(검색할 수 있는 형태로 변환)
   
  - 3. 검색(querying): 사용자가 검색창에 검색어나 질의를 입력하면, 검색엔진은 색인화 된 데이터를 기반으로 해당 검색어와 관련된 웹페이지를 찾아 검색 결과를 제공
   
- robots.txt 파일은 권고사항의 일종(국제적 규약)
- 크롤러가 어떤 페이지를 인덱싱 해야하고, 어떤 부분은 접근하면 안되는지에 대한 설명을 명시
- 목적:
  - 인증 페이지, 관리자 페이지, 민감한 정보가 있는 페이지에 대한 노출이 되지 않도록 차단
  - 서버가 크롤러에 의해 과도하게 접근되어 부하가걸리는 것을 방지(트래픽 관리)
  - 검색엔진이 원하지 않는 페이지는 인덱싱이 되지 않도록 조절
- robots exclusion protocol (로봇 제외 표준) 국제적 규약에 기반하여 사용(강제력이 없음)
- 보안으로써의 효력이 없음(url이 그대로 노출)
- robots.txt는 누구나 확인이 가능(공개되어 있는 파일)

```python
requests.get('https://naver.com/robots.txt').text
```

- **구성 요소**
- user-agent: 크롤러의 이름을 지정하는 곳
  - *: 해당 페이지를 접속하는 모든 유저에게 아래와 같은 규칙이 적용 됨
 
- disallow: 크롤러가 접근하지 못하게 차단된 경로(지정된 url크롤링 금지)
  - /: 해당 url을 포함한 모든 하위 페이지에 대한 접근을 금지
 
- allow: disallow로 금지된 url경로 중에서도 접근을 허용하는 예외적 경로
- sitemap: 사이트 맵 파일의 url지정, 검색엔진이 사이트의 구조를 더 쉽게 알도록 작성

**2. Beautifulsoup (bs4)**

- 파이썬을 사용하여 웹페이지를 긁어오는(scraping)작업을 쉽게 할 수 있도록 해주는 라이브러리
- html과 xml파일을 파싱(parsing)하여, 복잡한 정보에서 원하는 데이터를 추출, 수정, 검색할 수 있게 도움
  - html: 웹페이지를 생성하기 위한 markup 언어
    - 웹 페이지의 구조를 정의하고, 웹 브라우저에 표시되는 콘텐츠를 서술하는 데 사용
   
  - xml(extensible markup language): 데이터를 저장하고 전송하기 위한 markup언어
 
    - 웹 페이지의 구조를 전송하고 그 구조를 파악 및 표현하기 위해 사용
   
  - xml은 사용자가 필요에 따라 특정 태그를 직접 정의, html은 정해져 있는 규칙을 사용
 
  - latex: 과학 및 수학분야에서 문서를 작성하기 위한 markup언어

## 12-3. Web Table 데이터 수집

- 데이터 코스피 데이터를 가져와 분석
- https://finance.naver.com/sise/sise_market_sum.naver
- pd.read_html() 함수를 이용해서 html내 table형식으로 구성된 자료를 수집

```python
n = 7
total_list = []
count = 0

one_dic = {0: "", 1: "one", 2: "two", 3: "three", 4: "four", 5: "five", 6: "six", 7: "seven", 8: "eight", 9: "nine"}
ten_dic = {2: "twenty", 3: "thirty", 4: "forty", 5: "fifty", 6: "sixty", 7: "seventy", 8: "eighty", 9: "ninety"}
hund_dic = {1: "onehundred", 2: "twohundred", 3: "threehundred", 4: "fourhundred", 5: "fivehundred", 6: "sixhundred", 7: "sevenhundred", 8: "eighthundred", 9: "ninehundred"}
t = ['ten', 'eleven', 'twelve', 'thirteen', 'fourteen', 'fifteen', 'sixteen', 'seventeen', 'eighteen', 'nineteen']
answer = {i + 10: t[i] for i in range(10)}
answer[0] = ""
for num in range(1, 1000):
    if num // 10 == 0:
        answer[num] = one_dic[num]
    elif num < 20:
        continue
    elif num // 100 == 0:
        ten = num // 10
        one = num % 10
        answer[num] = ten_dic[ten] + one_dic[one]
    else:
        hund = num // 100
        answer[num] = hund_dic[hund] + answer[num % 100]

for _ in range(n):
    line = input()
    if line != "$":
        count += len(line)
    total_list.append(line)

for num in range(1, 1000):
    able = answer[num]
    length = len(able)
    if num == count + length:
        ans = num
        break

for word in total_list:
    if word == "$":
        print(answer[ans], end = " ")
    else:
        print(word, end = " ")
```
