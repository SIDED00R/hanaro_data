```python
# 라이브러리 호출
from bs4 import BeautifulSoup # 웹에서 받은 html 정보를 파이썬에서 자료구조로 변호나 및 사용
import requests # 웹으로부터 요청받아 자료를 다운로드 받을 때 사용
import time # 크롤링 수행 시 코드 수행 시간에 대한 간격을 부여
import random # 시간의 간격을 무작위로 부여하기 위해 난수 생성
import pickle # 현재까지 수집된 자료나 정보를 저장, 불러오기
```

```python
# 수집 url 선언
url = "https://search.daum.net/search?w=tot&rtmaxcoll=LOT&DA=LOT&q=1152%ED%9A%8C%20%EB%A1%9C%EB%98%90%EB%8B%B9%EC%B2%A8%EB%B2%88%ED%98%B8"

# 해당 url에 웹 정보를 요청
url_data = requests.get(url).text

# 문자로 받은 웹 정보를 html 자료구조로 변환
html = BeautifulSoup(url_data)

# 각 번호 데이터를 추출
num_list_tag = html.find('div', class_ = 'lottonum').find_all('span')

# 리스트 컴프리헨션을 사용해 숫자 데이터만 추출
[x.get_text() for x in num_list_tag]
```

```python
# 1. 로또(로또 6/45) 1회부터 현재 회차까지 로또 복권 1등 당첨번호를 수집하고, 
# Excel 저장하는 코드를 작성 (네이버/다음 웹사이트 이용)
# 반복문을 활용하여, 1회차에서 5회차 까지 데이터를 수집 -> 데이터프레임 형태로 변환
all_number = [] # 수짐된 모든 로또번호를 하나의 리스트에 추가
for i in range(1, 6): # 1회부처 5회까지 로또 번호를 수집
    # 수집 url 선언
    url = f"https://search.daum.net/search?w=tot&rtmaxcoll=LOT&DA=LOT&q={i}%ED%9A%8C%20%EB%A1%9C%EB%98%90%EB%8B%B9%EC%B2%A8%EB%B2%88%ED%98%B8"
    
    # 해당 url에 웹 정보를 요청
    url_data = requests.get(url).text
    
    # 문자로 받은 웹 정보를 html 자료구조로 변환
    html = BeautifulSoup(url_data)
    
    # 각 번호 데이터를 추출
    num_list_tag = html.find('div', class_ = 'lottonum').find_all('span')
    
    # 리스트 컴프리헨션을 사용해 숫자 데이터만 추출
    all_number.append([x.get_text() for x in num_list_tag] + [f"{i} 회차"])
all_number
```

```python
# 2. 위의 문제에서 실시간으로 데이터를 저장하게 끔 작성 / 코드를 멈춘 뒤, 
# 다시 실행 했을 때, 수집이 안되 회차 부터 이어서 수집하는 코드를 작성 
# (예를 들어 1회부터 10회까지 수집을 하다가 멈췄을 때, 해당 코드를 다시 실행하면, 11회차 부터 수집이 되 게끔 작성)

# 현재 사용하던 리스트를 파일 형태로 저장
pickle.dump(all_number, open('lotto_number.dat', 'wb'))

# 기존에 저장된 파일을 호출
all_number = pickle.load(open('lotto_number.dat', 'rb'))
for i in range(len(all_number) + 1, 10):
    # 수집 url 선언
    url = f"https://search.daum.net/search?w=tot&rtmaxcoll=LOT&DA=LOT&q={i}%ED%9A%8C%20%EB%A1%9C%EB%98%90%EB%8B%B9%EC%B2%A8%EB%B2%88%ED%98%B8"
    
    # 해당 url에 웹 정보를 요청
    url_data = requests.get(url).text
    
    # 문자로 받은 웹 정보를 html 자료구조로 변환
    html = BeautifulSoup(url_data)
    
    # 각 번호 데이터를 추출
    num_list_tag = html.find('div', class_ = 'lottonum').find_all('span')
    
    # 리스트 컴프리헨션을 사용해 숫자 데이터만 추출
    all_number.append([x.get_text() for x in num_list_tag] + [f"{i} 회차"])

all_number
```

```python
# 코드의 진행 상황을 시각화
from tqdm import tqdm

try:
    all_number = pickle.load(open('lotto_number.dat', 'rb'))
except:
    all_number = []

for i in tqdm(range(len(all_number) + 1, 1168)):
    # 수집 url 선언
    url = f"https://search.daum.net/search?w=tot&rtmaxcoll=LOT&DA=LOT&q={i}%ED%9A%8C%20%EB%A1%9C%EB%98%90%EB%8B%B9%EC%B2%A8%EB%B2%88%ED%98%B8"
    
    # 해당 url에 웹 정보를 요청
    url_data = requests.get(url).text
    
    # 문자로 받은 웹 정보를 html 자료구조로 변환
    html = BeautifulSoup(url_data)
    
    # 각 번호 데이터를 추출
    num_list_tag = html.find('div', class_ = 'lottonum').find_all('span')
    
    # 리스트 컴프리헨션을 사용해 숫자 데이터만 추출
    all_number.append([x.get_text() for x in num_list_tag] + [f"{i} 회차"])
    pickle.dump(all_number, open('lotto_number.dat', 'wb'))

all_number
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
```

```python
# 웹 드라이버를 실행하여 웹 페이지를 열기
driver = webdriver.Chrome()
```

```python
# 웹 드라이버에 의해 열린 웹 페이지를 닫기
driver.quit()
```

```python
# 이마트 몰 웹 페이지를 웹드라이버를 활용해 접속
driver = webdriver.Chrome()
driver.get('http://emart.ssg.com/') # 특정 url접속
```

```python
# 페이지가 불러올 때 까지 다음 코드를 지연
driver.implicitly_wait(10) # 10초동안 코드를 지연, 10초 이전에 요청한 페이지의 정보가 다 날아온다면 바로 다음 코드를 실행

# 크롤러 상에 띄워진 화면에서 '메뉴' 버튼에 해당하는 태그를 찾아
# 해당 태그의 xpath 형식의 자료를 이용하여 클릭
driver.find_element(By.XPATH, '//*[@id="skip_gnb"]/div/div[1]/button/span[1]').click()
```

```python
# 상품명에 해당하는 태그를 찾아 데이터를 추출
driver.find_element(By.XPATH, '//*[@id="ty_thmb_view"]/ul/li[1]/div/a/div[1]').text
```

```python
for i in range(1, 85):
    name = driver.find_element(By.XPATH, f'//*[@id="ty_thmb_view"]/ul/li[{i}]/div/a/div[1]').text
    print(name)
```

```python
for i in range(1, 10):
    name = driver.find_element(By.XPATH, f'//*[@id="ty_thmb_view"]/ul/li[{i}]/div/a/div[2]/div[1]/div/em').text
    print(name)
```

```python
# 상품명과 상품 금액에 대한 정보를 반복문을 이용하여 수집
product_name_list = [] # 반복문을 통해 데이터를 추가할 리스트를 구성
product_price_list = []
for i in range(1, 10):
    # 반복을 수행하다가, 해당 태그에 대한 정보가 없다면, 예외처리를 통해 건너뛰고 이어서 수집
    try:
        product_name = driver.find_element(By.XPATH, f'//*[@id="ty_thmb_view"]/ul/li[{i}]/div/a/div[1]').text
        product_price = driver.find_element(By.XPATH, f'//*[@id="ty_thmb_view"]/ul/li[{i}]/div/a/div[2]/div[1]').text
        product_name_list.append(product_name)
        product_price_list.append(product_price)
    except:
        pass

pd.DataFrame({'상품명': product_name_list, '상품가격': product_price_list})
```
