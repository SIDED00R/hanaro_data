```python
# 13. '방송시작시간'을 날짜형식으로 변환하여, '방송연도'/'방송월'/'방송일'/'방송요일'항목을 생성하시오.

df1['year'] = pd.to_datetime(df1['방송시작시간']).dt.year
df1['month'] = pd.to_datetime(df1['방송시작시간']).dt.month
df1['day'] = pd.to_datetime(df1['방송시작시간']).dt.day
df1['week'] = pd.to_datetime(df1['방송시작시간']).dt.day_name()
df1[['year', 'month', 'day', 'week', '방송시작시간']].head(2)
```

```python
# 14. '방송요일'항목에서 주말과 주중을 구분하여, '주말'과 '주중'의 '상품주문수량'과 '상품주문금액'의 평균의 차이를 계산하시오.

day_name_list = df1['week'].unique()
weekend_list = ['주중', '주말', '주말', '주중', '주중', '주중', '주중']
weekend_dict = dict(zip(day_name_list, weekend_list))
df1['주말구분'] = df1['week'].replace(weekend_dict)
df1['주말구분'].value_counts()
```

```python
cond1 = df1['주말구분'] == '주중'
cond2 = df1['주말구분'] == '주말'
df1_1 = df1.loc[cond1]
df1_2 = df1.loc[cond2]
```

```python
# 15. '배송방식'과 '매입형태'간의 연관성이 있는지 확인하고자 한다.
# - Contingency Table을 계산하고, 두 항목간 독립성 검정을 수행하시오.

c1 = pd.crosstab(df1['배송방식'], df1['매입형태'])
c1
```

```python
# 귀무가설: 배송방식과 매입형태 서로 독립(연관이 없다)
# 대립가설: 배송방식과 매입형태 서로 독립이 아니다(연관이 있다)
stats.chi2_contingency(c1)
# p.value(귀무가설이 참일 확률) < 5% 유의수준 -> 귀무가설 기각 -> 대립가설 참 -> 배송방식과 매입형태 서로 독립이 아니다.
```

```python
# 16. '방송구분'에서 '생방송'과 '녹화방송'의 매출을 비교하고자 한다. '생방송'과'녹화방송'의 '상품주문금액'의 평균의 차이가 있는지 검정하시오.
# - 가설 수립 및 결론 작성

# 귀무가설: 해당 데이터의 분포는 정규분포를 따른다.
# 대립가설: 해당 데이터의 분포는 정규분포를 따르지 않는다.
cond1 = df1['방송구분'] == '생방송'
cond2 = df1['방송구분'] == '녹화방송'
df1_live = df1.loc[cond1]
df1_record = df1.loc[cond2]
print(stats.normaltest(df1_live['상품주문금액']))
print(stats.normaltest(df1_record['상품주문금액']))
# p.value -> 대립가설 참 -> 정규분포를 따르지 않는다.
```

```python
# 생방송과 녹화방송에 따른 상품주문 금액의 중앙 값의 차이가 있는가?
# 귀무가설: 생방송과 녹화방송에 따른 상품주문금액의 중앙값의 차이는 없다.
# 대립가설: 생방송과 녹화방송에 따른 상품주문금액의 중앙값의 차이가 있다.

stats.ranksums(df1_live['상품주문금액'], df1_record['상품주문금액'])
# 대립가설 참 -> 중앙값의 차이가 있다.
```

```python
# 17. '상품목표주문금액'에서 '상품주문금액'의 차이를 계산해, 
# 목표를 달성하지 못한 방송에 대해서는 0 / 목표를 달성한 방송은 1값으로 '목표달성여부' 변수를 생성하시오.

df1['목표차액'] = df1['상품주문금액'] - df1['상품목표주문금액']
cond1 = df1['목표차액'] >= 0
df1.loc[cond1, '목표달성여부'] = 1
df1.loc[~cond1, '목표달성여부'] = 0
df1['목표달성여부'].value_counts()
```

# 지도학습(특성공학 적용)

```python
# 목표변수와 설명변수를 분할
y = df1['목표달성여부']
x = df1[['소요분', '방송구분', '판매단가', 'ARS금액', '수수료율', 'week', 'day']]
x.head(2)
```

```python
# 학습 데이터와 검증 데이터 분할(학습 성능과 일반화 성능을 모두 측정)
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=1234)
```

```python
# pipe line (특성공학 + 학습)
# 문자(범주형) 항목과 연속형 항목을 구분하여 특성공학 기법을 적용
from sklearn.compose import make_column_transformer # 파이프라인을 병렬로 분할
from sklearn.pipeline import make_pipeline #파이프라인 구성
# 숫자데이터(결측값 단순대치 -> 스케일링) 문자데이터(결측값 단순대치 -> 인코딩)
from sklearn.preprocessing import MinMaxScaler # 모든 숫자 데이터를 최댓값1, 최솟값0 변환
from sklearn.preprocessing import OneHotEncoder # 문자 항목을 1,0으로 구분할 수 있는 column으로 변환
from sklearn.impute import SimpleImputer # 결측값에 대한 단순 대치
# 학습 알고리즘에 의한 학습 수행
from sklearn.tree import DecisionTreeClassifier
# 평가
from sklearn.metrics import classification_report # 분류 모델 평가 지표
```

```python
# 숫자 데이터가 처리될 파이프를 구성
numeric_pipe = make_pipeline(SimpleImputer(strategy='most_frequent'), MinMaxScaler())
numeric_pipe # 결측값 중앙값으로 대치 -> 스케일링
```

```python
# 문자 데이터가 처리될 파이프를 구성
category_pipe = make_pipeline(SimpleImputer(strategy='most_frequent'), OneHotEncoder())
category_pipe # 결측값 최빈값으로 대치 -> 인코딩
```

```python
# 숫자 파이프와 문자 파이프라인을 묶어서 데이터가 처리될 특성공학 파트를 구성
prepro_pipe = make_column_transformer((numeric_pipe, ['소요분', '판매단가', 'ARS금액', '수수료율', 'day']),
                       (category_pipe, ['방송구분', 'week']))
prepro_pipe
```

```python
# 특성공학 + 학습
pipe_model = make_pipeline(prepro_pipe, DecisionTreeClassifier())
pipe_model.fit(x_train, y_train)
```

```python
# 평가 수행
y_train_pred = pipe_model.predict(x_train)
y_test_pred = pipe_model.predict(x_test)
```

```python
# 평가 실시
print(classification_report(y_train, y_train_pred))
```

```python
print(classification_report(y_test, y_test_pred))
```

### 교차 검증 cross validation

- 학습 단계에서 일반화가 더욱 잘 수행되기 위해, 학습데이터를 여러 단계로 나누어 교차로 검증하는 학습 기법
- Train Set내에서 데이터를 train set과 validation set으로 구분하여, 반복적으로 validation set을 바꿔가며 학습
- 특성이 강한 데이터에 학습이 매몰되지 않고, 다양한 특성이 반영되어 학습이 수행될 수 있다.
- 과적합을 어느정도 방지하며, 각 설명변수들의 특성을 골고루 반영한 모델들을 도출
- 교차검증을 활용 했을 때, 학습 데이터 구조
  - train set (validation set + train set)
  - test set
- k-fold cross validation: 특정 k개수만큼 학습 데이터를 나누어, 교차로 검증하는 데이터 셋(validation set)을 바꿔가며 학습을 수행
  - 1000개 train set -> 5-fold cv(200개 씩 데이터를 분할하여 학습)
    - 0 ~ 200번 데이터 validation, 나머지 800개를 이용해 학습
    - 200 ~ 400번 데이터 vadlidation, 나머지 800개를 이용해 학습
    - ... 5번 반복
 
- stratified k-fold cross validation: 특정 k개수 만큼 학습데이터를 나누는데, 목표변수의 클래스의 비율을 고려하여 k개수 만큼 분할
  - 1000개 train set(y 정상 950, 이상 50)
    - 0 ~ 200번 데이터 validation(y - 이상 10개), 나머지 800개(y - 이상 40개)
    - ... 5번 반복
 
- fold 수가 많다고 해서 성능이 좋은 모델이 도출되지 않는다.
- 일반적으로 3 ~ 5회 정도의 fold cv 수행 실시
- fold수가 너무 많은 경우엔, validation set 데이터의 수가 너무 적어, cv모델성능 신뢰성 문제가 발생

```python
# 교차검증 적용
from sklearn.model_selection import GridSearchCV
grid_model = GridSearchCV(pipe_model, cv =5, param_grid={}, scoring='accuracy')
# scoring: 어떤 평가지표를 이용하여 최적으 모델 선택할지 결정
grid_model.fit(x_train, y_train)
```

```python
#교차 검증 후 , 가장 성능이 좋은 모델을 도출
best_model = grid_model.best_estimator_
best_model
```

### 하이퍼 파라미터 튜닝(매개변수 튜닝)

- hyperparameter: model을 학습시키기 전 학습하면서 발생하는 학습 아고리즘내 세팅되어 있는 매개변수
- hyperparamter tuning: 학습 알고리즘에 내장된 매개변수를 사용자가 사전에 조절하여, overfitting을 방지, 학습과 일반화 성능을 올리는 기법(각 알고리즘 내 존재하는 hyperparameter를 사용자가 알고 있어야함)
  - random search: 매개변수의 여러가지 조합을 무작위로 구성하여 학습
  - grid search: 모든 가능한 범위의 매개변수 조합에 대해 학습
  - bayesian optimization search: 베이지안 최적화 기법을 이용하여, 매개변수의 조합을 gaussian process를 이용해 모델링하여, 가장 불확실성이 낮은 매개변수를 찾아 학습

```python
hyperparameter = {'decisiontreeclassifier__max_depth': range(5, 11)} # 5개의 경우의 수 부여
grid_model = GridSearchCV(pipe_model, cv = 5, param_grid=hyperparameter, scoring='accuracy') # 교차검증 5회 * 하이퍼파라미터 5개 = 25번의 모델 생성
grid_model.fit(x_train, y_train)
best_model = grid_model.best_estimator_
```

```python
# best model 평가 수행
y_train_pred = best_model.predict(x_train)
y_test_pred = best_model.predict(x_test)

print(classification_report(y_train, y_train_pred))
print(classification_report(y_test, y_test_pred))
```

**bayesian optimization search**

- 처음엔 random한 hyperparameter를 넣고 시작하여, 이전의 학습 결과를 바탕으로 다음 학습에서 탐색할 hyperparameter를 선택하는 확률 모델을 이용하여 최적의 hyperparameter도출
- 성능이 좋게 나올 수 있는 하이퍼파라미터를 지능적으로 접근하여, grid search 기법에 비해 연산량이 더 낮고, 더 정확한 모델을 찾을 수 있는 기법
- 지능적으로 접근(획득함수에 의한 확률 모형 -> 가장 불확실성이 낮은 확률로 파라미터를 탐색)
  - 획득 함수: 최적화할 함수(모델성늘)에 대해 더 많은 정보가 필요한지, 아니면 현재까지의 정보가 충분한 하이퍼파라미터를 도출할 수 있을 지 결정하는 함수
    - 모델의 성능이 가장 높아지는 것
    - 성능이 높아질 가능성이 높은 파라미터를 탐색하는 함수
  - 획득 함수:
    - expected improvement function(ei, 기대 개선 함수)
    - 다음으로 선택할 하이퍼파라미터에서 가댓값을 계산한 다음, 성능이 얼마나 개선될지를 예측하여 그 값이 최대가 되는 방향으로 하이퍼파라미터를 탐색
    - probability of imporvement(pi, 개선 확률 함수)
    - 현재까지의 최적성능을 개선할 확률이 최대가 되는 하이퍼파리미터를 탐색
    - upper confidence bound function(ucb, 상한 신뢰 구간 함수)
    - 기댓값과 불확실성을 결합한 값을 이용하여 하이퍼파라미터를 탐색
    - 특정 하이퍼파라미터에서 예상 성능(기댓값)과 불확실성(표준편차)을 더한 값을 최대화하는 방향으로 탐색
    - 아직 탐색해보지 않은 영역을 적극적으로 탐색하여 성능을 도출
